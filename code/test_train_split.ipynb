```python
# Import data

df = pd.read_csv(path+"/data/joint_data_collection_KategorieVectorized.csv", index_col="Unnamed: 0")

y_cols = [
    "Category_ATV",
    "Category_Allround",
    "Category_Classic",
    "Category_Cross / motocross",
    "Category_Custom / cruiser",
    "Category_Enduro / offroad",
    "Category_Naked bike",
    "Category_Scooter",
    "Category_Sport",
    "Category_Sport touring",
    "Category_Super motard",
    "Category_Touring",
    "Category_Unspecified category"
]

# y: Labels
y = df[y_cols]

# X: Features (alles au√üer Kategorien)
X = df.drop(columns=y_cols)

```

```python
# Test-Train-Split

df_random = df.sample(frac=1, random_state=42).reset_index(drop=True)

df_activation_results = df_random.iloc[:10]
df_rest = df_random.iloc[10:]

train_size = int(len(df_rest) * 0.80)
df_train = df_rest.iloc[:train_size]
df_test = df_rest.iloc[train_size:]

df_activation = df_activation_results.drop(y_cols, axis=1)

#print(df_train.shape)
#print(df_test.shape)
```
